```python
import spacy
from gramformer import Gramformer

def clean_text(text):
  nlp = spacy.load("en_core_web_sm")
  gf = Gramformer(models=1, use_gpu=False)  # GPU optional


  doc = nlp(text)
  sentences = [sent.text for sent in doc.sents]

  from pysbd import Segmenter
  segmenter = Segmenter(language="en", clean=False)
  sentences = segmenter.segment(text)

  corrected_sentences = []
  for sent in sentences:
      corrected = gf.correct(sent)
      corrected_sentences.append(list(corrected)[0])

  return " ".join(corrected_sentences)
```

```python
from transformers import pipeline
import re

# Clean text
def summary(text):
  cleaned_text = clean_text(text)

  # Remove special characters and extra whitespace
  cleaned_text = re.sub(r'[^\w\s.,!?]', '', cleaned_text).strip()

  # Initialize summarizer (fallback to smaller model)
  try:
      summarizer = pipeline(
          "summarization",
          model="sshleifer/distilbart-cnn-12-6",
          device="cpu"
      )
  except Exception as e:
      print("Model load error:", e)
      summarizer = None

  # Generate summary if text is valid
  if not cleaned_text:
      print("Error: Input text is empty.")
  elif len(cleaned_text) < 30:
      print("Text too short. Original:", cleaned_text)
  elif summarizer:
      try:
          summary = summarizer(
              cleaned_text,
              max_length=130,
              min_length=30,
              truncation=True,
              do_sample=False
          )
          print("Summary:", summary[0]['summary_text'])
      except Exception as e:
          print("Summarization failed. Error:", e)
          print("Falling back to extractive summarization...")
          # Add the spaCy/NLTK extractive method here
  else:
      print("Summarizer not available.")
  return summary[0]['summary_text']
```

```python
from transformers import pipeline

classifier = pipeline("zero-shot-classification", model="facebook/bart-large-mnli")

def classify_pros_cons(text):
    sentences = [s.strip() for s in text.split(". ") if s.strip()]
    results = {"Pros": [], "Cons": []}
    for sent in sentences:
        output = classifier(sent, candidate_labels=["pro", "con", "neutral"])
        label = output["labels"][0]
        if label == "pro":
            results["Pros"].append(sent)
        elif label == "con":
            results["Cons"].append(sent)
    return results
```
